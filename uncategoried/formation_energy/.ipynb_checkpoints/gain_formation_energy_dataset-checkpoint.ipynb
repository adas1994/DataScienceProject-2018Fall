```python
from pymatgen import Composition, Element

formation_energy_dictionary={}

inp=data
#inp['T'] = inp['T']/inp['N']
for i in range(len(inp['T'])):
    #print('---', i,'---')
    #print('Formation energy (eV/atom): ', inp['T'][i])
    #print('Atoms: ')

    compound_dictionary={}
    for element_number in inp['Z'][i]:
        element=str(Element.from_Z(element_number))
        if element in compound_dictionary:
            compound_dictionary[element]+=1
        else:
            compound_dictionary[element]=1
    compound_string=""
    for element in compound_dictionary:
        compound_string+=element+str(compound_dictionary[element])
    formation_energy_dictionary[compound_string]=inp['T'][i]
```
